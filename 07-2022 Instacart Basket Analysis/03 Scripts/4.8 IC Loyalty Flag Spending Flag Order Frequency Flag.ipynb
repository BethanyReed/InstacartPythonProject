###  3. Create Loyalty Flag for Existing Customers

```python
#Create new column with max orders for each user "max_order"
```

```python
ords_prods_merge['max_order'] = ords_prods_merge.groupby(['user_id'])['order_number'].transform(np.max)
```

```python
ords_prods_merge.head()
```

```python
#Create Loyalty Flag based on criteria:
#If the maximum orders the user has made is over 40, then the customer will be labeled a “Loyal customer.”
#If the maximum orders the user has made is over 10 but less than or equal to 40, then the customer will be labeled a “Regular customer.”
#If the maximum orders the user has made is less than or equal to 10, then the customer will be labeled a “New customer.”
```

```python
ords_prods_merge.loc[ords_prods_merge['max_order'] > 40, 'loyalty_flag'] = 'Loyal customer'
ords_prods_merge.loc[(ords_prods_merge['max_order'] <= 40) & (ords_prods_merge['max_order'] > 10), 'loyalty_flag'] = 'Regular customer'
ords_prods_merge.loc[ords_prods_merge['max_order'] <= 10, 'loyalty_flag'] = 'New customer'
```

```python
# Check counts for loyalty flag
```

```python
ords_prods_merge['loyalty_flag'].value_counts(dropna = False)
```

```python
ords_prods_merge[['user_id', 'loyalty_flag', 'order_number']].head(60)
```

### 5. Create Spending Flag per user for avg price across all their orders

```python
# If the mean of the prices of products purchased by a user is lower than 10, then flag them as a “Low spender.”
# If the mean of the prices of products purchased by a user is higher than or equal to 10, then flag them as a “High spender.”
```

```python
#Create column of average price of price per user
```

```python
ords_prods_merge['mean_price'] = ords_prods_merge.groupby(['user_id'])['prices'].transform(np.mean)
```

```python
#Checking new column stats
```

```python
ords_prods_merge.head()
```

```python
ords_prods_merge[['user_id', 'mean_price', 'prices']].head(60)
```

```python
# Create Spending Flag based on criteria:
# Low Spender = lower than 10 for mean
# High Spender = higher than or equal to 10
```

```python
ords_prods_merge.loc[ords_prods_merge['mean_price'] < 10, 'spending_flag'] = 'Low Spender'
ords_prods_merge.loc[ords_prods_merge['mean_price'] >= 10, 'spending_flag'] = 'High Spender'
```

```python
#Check spending flag
```

```python
ords_prods_merge['spending_flag'].value_counts(dropna = False)
```

### 6. Create Order Frequency Flag per user based on how often they order

```python
#If the median of “days_since_prior_order” is higher than 20, then the customer should be labeled a “Non-frequent customer.”
#If the median is higher than 10 and lower than or equal to 20, then the customer should be labeled a “Regular customer.”
#If the median is lower than or equal to 10, then the customer should be labeled a “Frequent customer.”
```

```python
#Create new column order_frequency for user_id based on median of days_since_prior_order
```

```python
ords_prods_merge['order_frequency'] = ords_prods_merge.groupby(['user_id'])['days_since_prior_order'].transform(np.median)
```

```python
#Check new column
```

```python
ords_prods_merge.head()
```

```python
ords_prods_merge[['user_id', 'order_frequency', 'days_since_prior_order']].head(60)
```

```python
# Create oder__freq_flag
# Non-frquent customer - higher than 20
# Regular Customer - higher than 10 lower than or equal to 20
#Frequent Customer - lower than or equal to 10
```

```python
ords_prods_merge.loc[ords_prods_merge['order_frequency'] > 20, 'order_frequency_flag'] = 'Non Frequent Customer'
ords_prods_merge.loc[(ords_prods_merge['order_frequency'] > 10) & (ords_prods_merge['order_frequency'] <= 20), 'order_frequency_flag'] = 'Regular Customer'
ords_prods_merge.loc[ords_prods_merge['order_frequency'] <= 10, 'order_frequency_flag'] = 'Frequent Customer'
```

```python
# Check values
```

```python
ords_prods_merge['order_frequency_flag'].value_counts(dropna = False)
```

```python
ords_prods_merge[['user_id', 'order_frequency_flag', 'order_frequency']].head(60)
```
