### 4. Divide customers into geographic regions

```python
#Divide customers into geographic regions by northeast, midwest, south, west
```

```python
result = []

for value in instacart2['state']:
  if value in ('Maine','New Hampshire','Vermont','Massachusetts','Rhode Island','Connecticut','New York','Pennsylvania','New Jersey'):
    result.append('Northeast')
  elif value in ('Wisconsin','Michigan','Illinois','Indiana','Ohio','North Dakota','South Dakota','Nebraska','Kansas','Minnesota','Iowa','Missouri'):
    result.append('Midwest')
  elif value in ('Delaware','Maryland','District of Columbia','Virginia','West Virginia','North Carolina','South Carolina','Georgia','Florida','Kentucky','Tennessee','Mississippi','Alabama','Oklahoma','Texas','Arkansas','Louisiana'):
    result.append('South')
  else:
    result.append('West')
```

```python
#Check results
```

```python
result
```

```python
instacart2['geographic_region'] = result
```

```python
instacart2['geographic_region'].value_counts(dropna=False)
```

```python
instacart2.groupby('geographic_region').agg({'prices': ['mean']})
```

### 6. Exclusion for customers with less than 5 orders

```python
#create exclusion flag column five_or_more with yes/no to include or not
```

```python
instacart2.loc[instacart2['max_order'] < 5, 'five_or_more'] = 'No'
instacart2.loc[instacart2['max_order'] >= 5, 'five_or_more'] = 'Yes'
```

```python
instacart2['five_or_more'].value_counts(dropna = False)
```

```python
#create new dataframe with only "yes" from exclusion flag five_or_more
```

```python
instacart_final =  instacart2[instacart2['five_or_more']=='Yes']
```

## Set price_range_loc

```python
instacart_final.loc[instacart_final['prices'] > 15, 'price_range_loc'] = 'High-range product'
```

```python
instacart_final.loc[(instacart_final['prices'] <= 15) & (instacart_final['prices'] > 5), 'price_range_loc'] = 'Mid-range product' 
```

```python
instacart_final.loc[instacart_final['prices'] <= 5, 'price_range_loc'] = 'Low-range product'
```
