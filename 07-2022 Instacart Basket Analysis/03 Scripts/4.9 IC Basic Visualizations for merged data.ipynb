### 7. Scatterplot age and income

```python
# Scatterpot to view connection between age and spending power
```

```python
sns.scatterplot(x = 'age', y = 'income',data = instacart)
```

### This chart is a scatterplot chart.  Across the bottom is age.  You would think that generally speaking as you got older you would have income/spending power.  But as you can see, below 300,000 income it doesn't seem to matter your age.  There is more correlation between 200,000 and above age 40, but not by a lot.  The other big diveder you can see is age 40 and anything above 400,000.  Just seems that that pay/income would generally have more to do with bumping up in your industry or becoming a manager/supervisor and therefore a higher income.  So you can see there is something to age and income, but it isn't a strong correlation where being older definitely means you have a higher income.  But you can see the HIGHEST income would indicate you are at least 40 or older
