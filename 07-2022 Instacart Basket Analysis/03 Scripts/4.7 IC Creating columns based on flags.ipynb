### 3. user defined function price_label

```python
#Define Function
```

```python
def price_label(row):

  if row['prices'] <= 5:
    return 'Low-range product'
  elif (row['prices'] > 5) and (row['prices'] <= 15):
    return 'Mid-range product'
  elif row['prices'] > 15:
    return 'High range'
  else: return 'Not enough data'
```

```python
#run function
```

```python
df_new['price_range'] = df_new.apply(price_label, axis=1)
```

```python
df_new.value_counts('price_range')
```

```python
df_new['prices'].max()
```

### 4. Using loc() function price_range_loc

```python
#set conditions
```

```python
df_new.loc[df_new['prices'] > 15, 'price_range_loc'] = 'High-range product'
```

```python
df_new.loc[(df_new['prices'] <= 15) & (df_new['prices'] > 5), 'price_range_loc'] = 'Mid-range product'
```

```python
df_new.loc[df_new['prices'] <= 5, 'price_range_loc'] = 'Low-range product'
```

```python
#Compare value counts
```

```python
df_new.value_counts('price_range_loc')
```

### 5. For Loops busiest_day

```python
#finding busiest day
```

```python
df_ords_prods_merge['orders_day_of_week'].value_counts(dropna = False)
```

```python
#0 is busiest day and is for Saturday, 4 is slowest which is Wednesday
```

```python
#run loop
```

```python
result = []

for value in df_ords_prods_merge["orders_day_of_week"]:
  if value == 0:
    result.append("Busiest day")
  elif value == 4:
    result.append("Least busy")
  else:
    result.append("Regularly busy")
```

```python
result
```

```python
df_ords_prods_merge['busiest_day']=result
```

```python
df_ords_prods_merge['busiest_day'].value_counts(dropna=False)
```

### 6. for loop busiest_days

```python
# 2 create busiest dayS column with 2 busiest days and 2 slowest days
```

```python
# as above, 2 busiests are 0 & 1, two slowest are 3 & 4
```

```python
result = []

for value in df_ords_prods_merge["orders_day_of_week"]:
  if value in (0,2):
    result.append("Busiest days")
  elif value in (3,4):
    result.append("Slowest days")
  else:
    result.append("Regular days")
```

```python
result
```

```python
df_ords_prods_merge['busiest_days']=result
```

```python
df_ords_prods_merge['busiest_days'].value_counts(dropna=False)
```

### 7. loc() function busiest hours of the day

```python
df_ords_prods_merge['order_hour_of_day'].value_counts(dropna = False)
```

```python
# Labels are Most orders, Average orders, and Fewest orders
# Most Orders = 9-17 (above 200,000)
# Average Orders = 7-8,18-23 (above 30,000 and below 200,000)
# Fewest Orders = 0-6 (below 30,000)
```

```python
df_ords_prods_merge.head()
```

```python
df_ords_prods_merge.loc[df_ords_prods_merge['order_hour_of_day'] <= 6, 'busiest_period_of_day'] = 'Fewest Orders'
```

```python
df_ords_prods_merge.loc[(df_ords_prods_merge['order_hour_of_day'] >= 9) & (df_ords_prods_merge['order_hour_of_day'] <= 17), 'busiest_period_of_day'] = 'Most Orders'
```

```python
df_ords_prods_merge.loc[(df_ords_prods_merge['order_hour_of_day'] >= 7) & (df_ords_prods_merge['order_hour_of_day'] <= 8), 'busiest_period_of_day'] = 'Average Orders'
```

```python
df_ords_prods_merge.loc[(df_ords_prods_merge['order_hour_of_day'] >= 18) & (df_ords_prods_merge['order_hour_of_day'] <= 23), 'busiest_period_of_day'] = 'Average Orders'
```
